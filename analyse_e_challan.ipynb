```python
import numpy as np
import pandas as pd
data=pd.read_csv("echallan_daily_data.csv")
print(data.info())
```

```python
df=data.describe()
```

```python
data['date']=pd.to_datetime(data['date'])
print(data.head(5))
print(data.isnull().sum())
```

```python
print(data.shape)
print(data["disposedAmount"].duplicated().sum()) # No Duplicates if date is , else if amount may there's
```

```python
data['year']=data['date'].dt.year
data.loc[(data['year']<2020),"Free Kardo"]= 0
data.loc[(data['year']>2020),"Free Kardo"]= 1
print((data['Free Kardo']==0).sum())
```

```python
print((data['pendingChallan']<53038.000000).sum())
```

```python
print((data['totalChallan']>174420).cumsum())
```

```python
data['disposed ratio']=data['disposedAmount']/data['totalChallan']
print(data.head(5))
```

```python
data['Collection Rate']=data['disposedAmount']/data['totalAmount']
print(data.head(5))
```

```python
data['avg_fine'] = data['totalAmount'] / data['totalChallan']

print(data.corr())
```

```python
data['pending_ratio']=data['pendingChallan']/data['disposedChallan']
print(data.head(5))
```

```python
data['Cumulative_challan']=data['pendingChallan'].cumsum()
print(data.head(5))
```

```python
data['Court ratio']=100* data['totalCourt']/data['totalChallan']
print((data['Court ratio']>50).sum())
```

```python
data['court_disposal_rate'] = data['disposedCourt'] / data['totalCourt']

print(data.head(5))
```

```python
print(data.corr())
```

```python
corr_matrix = np.corrcoef(data[['totalChallan',
                               'pendingChallan',
                               'totalAmount',
                               'totalCourt']].T)
print(corr_matrix)
```

```python
data['growth_rate'] = data['totalChallan'].pct_change() * 100
print(data['growth_rate'].head(10))
```

```python
mean = np.mean(data['totalChallan'])
std = np.std(data['totalChallan'])

outliers = data[np.abs(data['totalChallan'] - mean) > 3*std]
print(outliers)
```

```python

summary = {
    "Total Challans": data['totalChallan'].sum(),
    "Total Revenue": data['totalAmount'].sum(),
    "Avg Daily Challan": data['totalChallan'].mean(),
    "Max Revenue Day": data['totalAmount'].max()
}
print(summary)
```
